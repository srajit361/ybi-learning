```python
sn_features = sn[['Movie_Genre', 'Movie_Keywords', 'Movie_Tagline', 'Movie_Cast',
       'Movie_Director']].fillna('')
```

```python
sn_features.shape
```

```python
sn_features
```

```python
X = sn_features['Movie_Genre'] + ' ' + sn_features['Movie_Keywords'] + ' ' + ' ' + sn_features['Movie_Tagline'] + ' ' + sn_features['Movie_Cast'] + ' '  + sn_features['Movie_Director']
```

```python
X
```

```python
X.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
X=tfidf.fit_transform(X)
```

```python
X.shape
```

```python
print(X)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_score = cosine_similarity(X)
```

```python
Similarity_score
```

```python
Similarity_score.shape
```

```python
Favourite_Movie_Name = input('Enter your favourite movie name: ')
```

```python
all_movie_tittle_list = sn['Movie_Title'].tolist()
```

```python
import  difflib
```

```python
movie_recommendation = difflib.get_close_matches(Favourite_Movie_Name, all_movie_tittle_list)
print(movie_recommendation)
```

```python
close_match = movie_recommendation[0]
print(close_match)
```

```python
Index_of_close_match_movie = sn[sn.Movie_Title == close_match]['Movie_ID'].values[0]
print(Inddex_of_close_match_movie)
```

```python
Recommendation_score = list(enumerate(Similarity_score[Index_of_close_match_movie]))
```

```python
len(Recommendation_score)
```

```python
sorted_similar_movies = sorted(Recommendation_score, key = lambda x:x[1], reverse = True)
```

```python
print ('Top 30 movies recommended for you : \n')
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index = sn[sn.index==index]['Movie_Title'].values[0]
  if (i<31):
    print(i, '.', title_from_index)
    i+=1
```
